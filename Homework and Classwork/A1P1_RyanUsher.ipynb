```python
import numpy as np
import matplotlib.pyplot as plt
f = lambda w, x : w[0] + w[1] * x + w[2] * (x**2) + w[3] * (x**3 )
# sample 20 e q u all y spaced v al u e s between −2 and 2
dom = np.linspace(-2, 2 , 20)
w = np.random.rand(4)
val = f(w, dom) + np.random.randn(20)/2
valtruth = f(w, dom)

#####################
A = np.zeros((20,4))
A[:, 0] = 1
for row in range(20):
  for col in range(1,4):
    A[row, col] = np.power(dom[row], col)
######################

b = np.expand_dims(val, axis =1)
w_learned = np.squeeze(np.linalg.lstsq(A,b)[0])
w_learned
val_learned = f(w_learned, dom)
fig,ax = plt.subplots(1 , 1 )
ax.plot(dom, val , 'bx' )
ax.plot(dom, valtruth ,'-g' )
ax.plot(dom, val_learned, '--r')
ax.legend([' Noisy samples' ,' Ground truth', 'Learned model'])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Noisy sample s and ground truth function' )
```
