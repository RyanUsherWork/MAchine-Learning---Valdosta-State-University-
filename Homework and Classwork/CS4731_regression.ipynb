```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales = [368, 340, 665, 954, 331, 556, 376]
ads = [1.7, 1.5, 2.8, 5, 1.3, 2.2, 1.3]

sales = np.array(sales)
ads = np.array(ads)

plt.figure()
plt.scatter(ads, sales)
plt.xlabel('Ads')
plt.ylabel('Sales')

```

```python

def model(B0, B1):
  predictions = B0 + B1 * ads
  return predictions
  
def loss(B0, B1):
  term1 = model(B0, B1)
  term2 = sales - term1
  term3 = term2 ** 2
  term4 = np.sum(term3)
  return term4 / 7
```

```python
blue_loss = loss(300, 50.2)
green_loss = loss(200, 100)

print("Blue loss: ", blue_loss)
print("Green loss: ", green_loss)
```

```python
delta = 0.0001
B0 = 300
B1 = 50.2

losses = []
# coordinate descent
for i in range(10000):
  # update B0
  l1 = loss(B0, B1)
  l2 = loss(B0 - delta, B1)
  partial_wrt_B0 = (l2 - l1) / delta
  B0 = B0 + partial_wrt_B0*0.001 # learning rate
  # update B1
  l1 = loss(B0, B1)
  l2 = loss(B0, B1 - delta)
  partial_wrt_B1 = (l2 - l1) / delta
  B1 = B1 + partial_wrt_B1*0.001 # learning rate
  #print("Loss: ", loss(B0, B1))
  losses.append(loss(B0, B1))

```

```python
plt.figure()
plt.plot(losses)
```
