```python
from skimage.io import imread
im = imread ('waterfall.jpg')
im2 = imread('nature-2-26-17.jpg')
im3 = imread('Skyscrapers_of_Shinjuku_2009_January_(revised).jpg')
im4 = imread('toronto-ontario.jpg')
im5 = imread('slideshow-new-vsu-arch.jpg')
im6 = imread('planningimage.jpg')
from skimage.transform import resize
```

```python
images = []
images.append(im)
images.append(im2)
images.append(im3)
images.append(im4)
images.append(im5)
images.append(im6)
```

```python
plt.imshow(images[5])
```

```python
image_small = []

for i in range (np.size(images)):
  image_small.append(resize(images[i], (images[i].shape[0] // 10, images[i].shape[1] // 10), anti_aliasing=True))
```

```python
from sklearn.cluster import KMeans
data = np.reshape(image_small[0], (np.shape(image_small[0])[0] * np.shape(image_small[0])[1], 3))

kmeans = KMeans(n_clusters=2, random_state=0).fit(data)

new_data = np.array(data)
for row in range(np.shape(new_data)[0]):
  #print(new_data[row,:])
  new_data[row, :] = kmeans.cluster_centers_[kmeans.labels_[row], :]
new_image = np.reshape(new_data, (images[0].shape[0] // 10, images[0].shape[1] // 10, 3))
plt.imshow(new_image)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(data)

new_data = np.array(data)
for row in range(np.shape(new_data)[0]):
  #print(new_data[row,:])
  new_data[row, :] = kmeans.cluster_centers_[kmeans.labels_[row], :]
new_image = np.reshape(new_data, (images[0].shape[0] // 10, images[0].shape[1] // 10, 3))
plt.imshow(new_image)
```
