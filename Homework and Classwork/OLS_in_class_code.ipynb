```python
import numpy as np

X = np.concatenate((np.ones((7, 1)), np.array([[1.7, 1.5, 2.8, 5, 1.3, 2.2, 1.3]]).T), axis=1)

Y = np.array([[368, 340, 665, 954, 331, 556, 376]]).T


beta = np.expand_dims(np.random.randn(2), axis=1)
#np.shape(beta)
Y_predicted = np.matmul(X, beta)
error = np.sum(np.power(np.abs(Y - Y_predicted), 2))
print("Error before optimization:", error)

from numpy.linalg import inv

####### equation on last slide on OLS 
beta_hat = np.matmul(np.matmul(inv(np.matmul(X.T,  X)), X.T), Y)

Y_predicted = np.matmul(X, beta_hat)
error = np.sum(np.power(np.abs(Y - Y_predicted), 2))
print("Error after optimization:", error)

import matplotlib.pyplot as plt

plt.figure()
x_viz = np.array([1.7, 1.5, 2.8, 5, 1.3, 2.2, 1.3])
y_viz = np.array([368, 340, 665, 954, 331, 556, 376])
plt.plot(x_viz, y_viz, 'rx')

domain = np.linspace(0, 5, 100)
values = beta_hat[0] + beta_hat[1] * domain
plt.plot(domain,values, '-g')
plt.xlabel("Ads")
plt.ylabel("Sales")



```
