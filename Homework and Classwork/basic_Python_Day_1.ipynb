```python
#floating point comparison
a = 0.1 + 0.2
b = 0.3
distance = abs(a - b)
print(a == b)
print(distance) #distance is not 0

#how to compare floats
if(distance < 0.000000001):
  print("Equal")
else:
  print("Not Equal")

a = 0.1 + 0.1
b = 0.2

print(a == b) #never use == to compare floats, since they are aproximations
```

```python
import numpy as np
import matplotlib.pyplot as plt 

#1D data 

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure();
ax = fig.add_subplot(1,1,1)

ax.plot(x,y)
ax.plot(x,z, marker="+", linestyle = "solid", linewidth = 1, markersize = 5)
ax.legend(["Sine", "Cosine"], loc = "upper right")

fig.show() #draws stuff immedaitely 

# code that does alot of work takes time to show in graph
```

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(1,1,1)
ax.clear()
ax.bar(x,y, width = 0.05)
```

```python
#2d Plots
[X, Y] = np.meshgrid(np.array([-2,-1,0,1,2]), np.array([-2,-1,0,1,2]))
print(X)
print()
print(Y)

x_flat = X.flatten()
y_flat = Y.flatten()
fig = plt.figure(figsize=(2,2))
ax = fig.add_subplot(1,1,1)
ax.clear()
ax.scatter(x_flat, y_flat)
```

```python
[X,Y] = np.meshgrid(np.linspace(-12,12,100), np.linspace(-12,12,100))

def fun(x,y):
  return np.sqrt(np.sin(x/4)**2 + np.cos(y/4)**2)

Z = fun(X,Y)

from mpl_toolkits.mplot3d import Axes3D
fig  = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X,Y,Z)

#Pyplot
```
